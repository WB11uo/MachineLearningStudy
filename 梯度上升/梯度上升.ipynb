```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.empty((100,2))
x[:,0]=np.random.uniform(0.,100.,size=100)
x[:,1]=0.75*x[:,0]+3.+np.random.normal(0,10.,size=100)
```

```python
plt.scatter(x[:,0],x[:,1])
plt.show()
```

```python
# demean
def demean(X):
    return X-np.mean(X,axis=0)
```

```python

X_demean=demean(x)
plt.scatter(X_demean[:,0],X_demean[:,1])
plt.show()
```
