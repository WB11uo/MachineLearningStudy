```python
import pandas as pd
```

```python
Date = pd.read_csv('completed.csv') #读取的不是本地文件，而是jupyter notebook服务器上的
```

```python
X=Date.loc[:,['面积','所属区']]
y=Date.loc[:,['价钱']]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
from sklearn.linear_model import  LinearRegression
linreg=LinearRegression()
model=linreg.fit(X_train,y_train)
```

```python
y_pred=linreg.predict(X_test)
```

```python
linreg.score(X_test,y_test) #评分标准与r2_score标准一样
```

```python
import matplotlib.pyplot as plt
plt.figure()
plt.plot(range(len(y_pred[1:300])),y_pred[1:300],'b',label="price_predict")
plt.plot(range(len(y_pred[1:300])),y_test[1:300],'r',label="price_test")
plt.legend(loc="upper right")
plt.show()
```

```python
from sklearn.metrics import r2_score
r2_score(y_test,y_pred)  
```
