```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.random.uniform(-3,3,size=100)
X=x.reshape(-1,1)
y=0.5*x**2+x+2+np.random.normal(0,1,100)
```

```python
plt.scatter(x,y)
plt.show()
```

```python
#使用线性回归
from sklearn.linear_model import LinearRegression
lin_reg=LinearRegression()
lin_reg.fit(X,y)
lin_reg.score(X,y)
```

```python
#均方误差
from sklearn.metrics import mean_squared_error
y_predict=lin_reg.predict(X)
mean_squared_error(y,y_predict)
```

```python
from sklearn.pipeline import  Pipeline
from sklearn.preprocessing import PolynomialFeatures
from sklearn.preprocessing import StandardScaler
def PolynomialRegression(degree):
    return Pipeline([
    ("poly",PolynomialFeatures(degree=degree)),
    ("std_scaler",StandardScaler()),
    ("lin_reg",LinearRegression())
    ])
```

```python
poly2=PolynomialRegression(degree=2)
poly2.fit(X,y)
```

```python
y2_predict=poly2.predict(X)
```

```python
mean_squared_error(y,y2_predict)
```

```python
poly3=PolynomialRegression(degree=10)
poly3.fit(X,y)
y3_predict=poly3.predict(X)
mean_squared_error(y,y3_predict)
```

```python
poly4=PolynomialRegression(degree=100)
poly4.fit(X,y)
y4_predict=poly4.predict(X)
mean_squared_error(y,y4_predict)
```

```python
plt.scatter(x,y)
plt.plot(np.sort(x),y4_predict[np.argsort(x)],color='r')
plt.show()
```
